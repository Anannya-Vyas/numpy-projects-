# Project 2: Pass/Fail Counter
What to build: A program that counts how many students passed an exam.


```python
import numpy as np
import sys
# using numpy to check pass fail easily
# user can exit anytime by typing -1


def readnum(msg):
    while True:
        try:
            n = input(msg)
            if n == "-1":
                print("PROGRAM EXITED BY USER")
                sys.exit()
            n = int(n)
            if n <= 0:
                print("ENTER NUMBER GREATER THAN ZERO")
            else:
                return n
        except ValueError:
            print("ENTER NUMBERS ONLY OR -1 TO EXIT")


def readmarks(msg, maxm):
    while True:
        try:
            m = input(msg)
            if m == "-1":
                print("PROGRAM EXITED BY USER")
                sys.exit()
            m = int(m)
            if m < 0 or m > maxm:
                print(f"MARKS MUST BE BETWEEN 0 AND {maxm}")
            else:
                return m
        except ValueError:
            print("ENTER NUMBERS ONLY OR -1 TO EXIT")


# main program

print("WELCOME TO PASS FAIL COUNTER")
print("TYPE -1 ANYTIME TO EXIT")

test = input("ENTER EXAM NAME (OR -1 TO EXIT): ")
if test == "-1":
    print("PROGRAM EXITED BY USER")
    sys.exit()

subcnt = readnum("ENTER TOTAL SUBJECTS: ")

subnames = []
submax = []
subpass = []

print("\nENTER SUBJECT DETAILS")

for i in range(subcnt):
    sname = input(f"\nENTER SUBJECT {i + 1} NAME (OR -1 TO EXIT): ")
    if sname == "-1":
        print("PROGRAM EXITED BY USER")
        sys.exit()

    tmarks = readnum(f"ENTER TOTAL MARKS FOR {sname}: ")

    while True:
        pmarks = readnum(f"ENTER PASS MARKS FOR {sname}: ")
        if pmarks > tmarks:
            print("PASS MARKS CANNOT BE GREATER THAN TOTAL MARKS")
        else:
            break

    subnames.append(sname)
    submax.append(tmarks)
    subpass.append(pmarks)


stucnt = readnum("\nENTER TOTAL STUDENTS: ")

passlist = []
faillist = []
failinfo = {}

for i in range(stucnt):
    print(f"\nENTER DETAILS FOR STUDENT {i + 1}")
    name = input("ENTER STUDENT NAME (OR -1 TO EXIT): ")
    if name == "-1":
        print("PROGRAM EXITED BY USER")
        sys.exit()

    marks = []
    fail = []

    for j in range(subcnt):
        m = readmarks(
            f"ENTER MARKS IN {subnames[j]} (0 TO {submax[j]}): ",
            submax[j]
        )
        marks.append(m)

        if m < subpass[j]:
            fail.append((subnames[j], m, subpass[j]))

    marr = np.array(marks)
    parr = np.array(subpass)

    if np.all(marr >= parr):
        passlist.append(name)
    else:
        faillist.append(name)
        failinfo[name] = fail


# final output

print("\nRESULT SUMMARY\n")
print("EXAM NAME:", test)
print("TOTAL STUDENTS:", stucnt)
print("STUDENTS PASSED:", len(passlist))
print("STUDENTS FAILED:", len(faillist))

print("\nPASSED STUDENTS:")
if passlist:
    for s in passlist:
        print("-", s)
else:
    print("NONE")

print("\nFAILED STUDENTS:")
if faillist:
    for s in faillist:
        print("-", s)
        for sub, m, pm in failinfo[s]:
            print(f"   {sub}: {m} (PASS MARKS {pm})")
else:
    print("NONE")
```
