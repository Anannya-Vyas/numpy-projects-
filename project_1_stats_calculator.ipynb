# What to build: A program that calculates basic statistics from a list of numbers.


Instructions:
1. Create a NumPy array with exactly 20 numbers (you can choose any numbers)

2. Calculate and print the following:  

o Mean (average)

o Median (middle value)

o Minimum value

o Standard deviation

3. Use NumPy functions: np.mean(), np.median(), np.min(), np.std()



```python
import numpy as np

print("Welcome to the NumPy Stats CalculatorðŸ¤–ðŸ¤–ðŸ¤–ðŸ¤–ðŸ¤–ðŸ¤–")
print("Please enter exactly 20 numbers\n")

# Taking input from user
num_list = []

while True:
    try:
        user_ip = input(" Enter numbers: ")
        num_list = list(map(float, user_ip.split()))
        count = len(num_list)

        if count < 20:
            print(f" You entered only {count} numbers. Please enter 20.")
            continue

        if count > 20:
            print(f" You entered {count} numbers. Please enter ONLY 20.")
            continue

        break  # exactly 20 â†’ success

    except ValueError:
        print("Invalid input! Enter numbers only.")

# Convert this list to NumPy array
numbers = np.array(num_list)

# Statistical calculations
# mean
mean_value = np.mean(numbers)
# median
median_value = np.median(numbers)
# min values
min_value = np.min(numbers)
# std deviation
std_deviation = np.std(numbers)

# Displaying results
print("\n Statistical Results")

print("->" * 80,'-','\n')


print("Mean (Average) of numbers TAKEN AS INPUT FROM USERS:\n", round(mean_value, 2), '\n')

print("Median of numbers TAKEN AS INPUT FROM USERS:\n", round(median_value, 2),'\n')

print("Minimum Value of numbers TAKEN AS INPUT FROM USERS:\n", min_value,"\n")
print("Standard Deviation of numbers TAKEN AS INPUT FROM USERS:\n", round(std_deviation, 2),'\n')
```
